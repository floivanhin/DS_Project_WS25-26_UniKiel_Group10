Code for experimenting with one game:

```python
import pandas as pd
import soccerdata as sd
import json

data = sd.WhoScored(leagues="GER-Bundesliga", seasons="2024/2025", headless=True)

schedule = data.read_schedule()

game_id = schedule.iloc[-1]["game_id"]

events = data.read_events(match_id=[game_id])

possession_changes = ["Tackle", "Interception", "BallRecovery"]

poss_gain_own_half = events[
    (events['type'].isin(possession_changes)) & 
    (events['outcome_type'] == 'Successful') & 
    (events['x'] < 50)
]

poss_gain_own_half
```

Data for all the times a team won possession in their own half:

```python
import pandas as pd
import soccerdata as sd

data = sd.WhoScored(leagues="GER-Bundesliga", seasons="2024/2025", headless=True)

game_ids = data.read_schedule()["game_id"].to_list()

events = data.read_events(match_id=game_ids)

possession_win = ['Tackle', 'Interception', 'BallRecovery']

own_half_wins = events[
    (events['type'].isin(possession_win)) & 
    (events['outcome_type'] == 'Successful') & 
    (events['x'] < 50)
]

own_half_wins

own_half_wins.to_csv("Possession Changes.csv",index=False)
```

Data on all the shots during the season:

```python
import pandas as pd
import soccerdata as sd

data = sd.WhoScored(leagues="GER-Bundesliga", seasons="2024/2025", headless=True)

game_ids = data.read_schedule()["game_id"].to_list()

events = data.read_events(match_id=game_ids)

shot_types = ['Goal', 'MissedShots', 'SavedShot', 'ShotOnPost']

shots = events[events["type"].isin(shot_types)]

shots

shots.to_csv("Shots.csv",index=False)
```

Data on all the substitutions during the season:

```python
import pandas as pd
import soccerdata as sd

data = sd.WhoScored(leagues="GER-Bundesliga", seasons="2024/2025", headless=True)

game_ids = data.read_schedule()["game_id"].to_list()

events = data.read_events(match_id=game_ids)

substitutions = events[events["type"]=="SubstitutionOff"]

substitutions

substitutions.to_csv("Substitutions.csv",index=False)
```
