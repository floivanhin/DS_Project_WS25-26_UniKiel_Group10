# RQ5: What is the relationship between payroll spending and league points?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import pearsonr
```

```python

df = pd.read_csv("budget.csv")
```

```python
r, p = pearsonr(df["budget(€)"], df["points"])
print(f"Pearson r = {r:.3f}, p = {p:.3f}")

sns.regplot(data=df, x="budget(€)", y="points", scatter_kws={"s":50})
for _, row in df.iterrows():
    if row["club"] in ["Bayern Munich","Bayer Leverkusen","Hoffenheim","Eintracht Frankfurt","Holstein Kiel"]:
        plt.text(row["budget(€)"], row["points"]+0.8, row["club"], fontsize=9)
plt.xlabel("Budget (€)")
plt.ylabel("League points")
plt.title("Budget vs Points")
plt.grid(True)
plt.show()

X = sm.add_constant(df["budget(€)"])
model = sm.OLS(df["points"], X).fit()
```
