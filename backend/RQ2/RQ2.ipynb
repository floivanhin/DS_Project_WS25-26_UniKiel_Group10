# RQ2: How does the matchday affect the amount of goals?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
```

```python
df = pd.read_csv("data_goals.csv")
```

```python
column = "total_goals_2020-2021"
plt.figure(figsize=(10,6))
plt.plot(df[column], marker="o")
plt.title("How does the matchday affect the amount of goals?")
plt.xlabel("matchday")
plt.ylabel("amount of goals")
plt.grid(True)
plt.tight_layout()
plt.show()
print("mean:", statistics.mean(df[column]))
print("median:", statistics.median(df[column]))
print("mode:", statistics.mode(df[column]))
print("min:", min(df[column]))
print("max:", max(df[column]))
print("range:", max(df[column]) - min(df[column]))
print("sum:", sum(df[column]))
```
