RQ2: How does the matchday affect the amount of goals?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("understat_schedule.csv")
```

```python
c = 0
b = 0
r = []
for i in range (306):
    b = b + df["away_goals"][i] + df["home_goals"][i]
    c += 1
    if c > 8:
        c = 0
        r = r + [b]
        b = 0
    else:
        pass
```

```python
plt.figure(figsize=(10,6))
plt.plot(r, marker="o")
plt.title("How does the matchday affect the amount of goals?")
plt.xlabel("matchday")
plt.ylabel("amount of goals")
plt.grid(True)
plt.tight_layout()
plt.show()
```
