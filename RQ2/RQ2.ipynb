# RQ2: How does the matchday affect the amount of goals?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
```

```python
df = pd.read_csv("RQ2.csv")
```

```python
plt.figure(figsize=(10,6))
plt.plot(df["total_goals"], marker="o")
plt.title("How does the matchday affect the amount of goals?")
plt.xlabel("matchday")
plt.ylabel("amount of goals")
plt.grid(True)
plt.tight_layout()
plt.show()
print("mean:", statistics.mean(df["total_goals"]))
print("median:", statistics.median(df["total_goals"]))
print("mode:", statistics.mode(df["total_goals"]))
print("min:", min(df["total_goals"]))
print("max:", max(df["total_goals"]))
print("range:", max(df["total_goals"]) - min(df["total_goals"]))
print("sum:", sum(df["total_goals"]))
```
